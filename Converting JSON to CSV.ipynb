## Converting Receipts JSON to readble CSV & transforming an additional rewards table

```python
# Reading the receipts json file using read_json and creating a dataframe
import json    
import pandas as pd
import datetime
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval

receipt_path = 'C:/Users/Parth Parsana/OneDrive/Desktop/Interview prep/Fetch reward/receipts.json'
receipt_df = pd.read_json(receipt_path, lines=True)
receipt_df = pd.DataFrame(receipt_df)
receipt_df
```

```python
# Checking Data Types
receipt_df.dtypes
```

```python
# Normalizing the Json format in required columns 
receipt_df['_id'] = pd.json_normalize(receipt_df['_id'],max_level=1)
receipt_df['createDate'] = pd.json_normalize(receipt_df['createDate'],max_level=1)
receipt_df['dateScanned'] = pd.json_normalize(receipt_df['dateScanned'],max_level=1)
receipt_df['modifyDate'] = pd.json_normalize(receipt_df['modifyDate'],max_level=1)
receipt_df['finishedDate'] = pd.json_normalize(receipt_df['finishedDate'])
receipt_df['pointsAwardedDate'] = pd.json_normalize(receipt_df['pointsAwardedDate'])
receipt_df['purchaseDate'] = pd.json_normalize(receipt_df['purchaseDate'])
```

```python
# Renaming column names
receipt_df.rename(columns = {'_id':'receiptId'}, inplace = True)
```

```python
receipt_df
```

```python
# Converting the columns to appropriate format
receipt_df['createDate'] = pd.to_datetime(receipt_df['createDate'],unit='ms')
receipt_df['dateScanned'] = pd.to_datetime(receipt_df['dateScanned'],unit='ms')
receipt_df['modifyDate'] = pd.to_datetime(receipt_df['modifyDate'],unit='ms')
receipt_df['finishedDate'] = pd.to_datetime(receipt_df['finishedDate'],unit='ms')
receipt_df['pointsAwardedDate'] = pd.to_datetime(receipt_df['pointsAwardedDate'],unit='ms')
receipt_df['purchaseDate'] = pd.to_datetime(receipt_df['purchaseDate'],unit='ms')
receipt_df['pointsEarned'] = pd.to_numeric(receipt_df['pointsEarned'])
receipt_df['purchasedItemCount'] = pd.to_numeric(receipt_df['purchasedItemCount'])
receipt_df = receipt_df.astype({'receiptId':'string','userId':'string','bonusPointsEarnedReason':'string','rewardsReceiptStatus':'string'})
receipt_df.head()
```

```python
# Creating another table for items table with rceeiptId and userId as foreign keys
dfrewardslist = receipt_df.filter(['rewardsReceiptItemList','userId','dateScanned', 'receiptId'], axis = 1)
rewards_df = dfrewardslist[~dfrewardslist['rewardsReceiptItemList'].isnull()]
rewards_df
```

```python
# Deleting receipt item list column from receipt table as it will be linked using items table
del receipt_df['rewardsReceiptItemList']
receipt_df.dtypes
```

```python
# Saving the receipt dataframe as a table(csv)
receipt_df.to_csv('C:/Users/Parth Parsana/OneDrive/Desktop/Interview prep/Fetch reward/receipts.csv',index=False)
```

```python
# Creating a new items table from the receipt item list column, using key : value pairings (Feature engineering)
items_df = pd.concat([pd.DataFrame(x) for x in rewards_df['rewardsReceiptItemList']], keys = rewards_df['receiptId']).reset_index()
items_df
```

```python
# Checking for sum of null values in every column
items_df.isna().sum()
```

```python
# Merging the reward table with receipt item list to get the new items table
item_df = pd.merge(rewards_df, items_df, how='left', on = 'receiptId')
del item_df['rewardsReceiptItemList']
item_df
```

```python
# Saving the items dataframe as a table(csv)
item_df.to_csv('C:/Users/Parth Parsana/OneDrive/Desktop/Interview prep/Fetch reward/items.csv',index=False)
```

## Converting Brands JSON to readble CSV

```python
# Reading the brands json file using read_json and creating a dataframe
brands_path = 'C:/Users/Parth Parsana/OneDrive/Desktop/Interview prep/Fetch reward/brands.json'
brands_df = pd.read_json(brands_path, lines=True)
brands_df = pd.DataFrame(brands_df)
brands_df
```

```python
# Checking data types
brands_df.dtypes
```

```python
# Normalizing the id field and renaming column names
brands_df['_id'] = pd.json_normalize(brands_df['_id'],max_level=1)
brands_df.rename(columns = {'_id':'brandId'}, inplace = True)
brands_df
```

```python
# Creating a cpg dataframe with brandId as index
dfcpg = brands_df.filter(['cpg', 'brandId'], axis = 1)
cpg_df = dfcpg[~dfcpg['cpg'].isnull()]
cpg_df
```

```python
# Creating a dataframe with cpg column values and brandId as primary key (Feature engineering)
cpgs_df = pd.concat([pd.DataFrame(x) for x in cpg_df['cpg']], keys = cpg_df['brandId']).reset_index()
cpgs_df
```

```python
# Renaming the column names
cpgs_df = cpgs_df.rename(columns = {'$id': 'cpg_Id', '$ref': 'cpf_Ref'})
cpgs_df
```

```python
# joining the brand and cpgs dataframe to create the final brand dataframe
b_df = pd.merge(brands_df, cpgs_df, how='left', on = 'brandId')
del b_df['cpg']
del b_df['level_1']
b_df
```

```python
# Checking column data types
b_df.dtypes
```

```python
# Filling NA values in topbrand with 0(false) and changing data types as per required
b_df['topBrand'] = b_df['topBrand'].fillna(0)
b_df = b_df.astype({'brandId':'string','cpg_Id':'string','cpf_Ref':'string','brandCode':'string','name':'string',
                    'category':'string','categoryCode':'string','topBrand':'bool'})
b_df
```

```python
# Checking data types
b_df.dtypes
```

```python
# Saving the brands dataframe as a table(csv)
b_df.to_csv('C:/Users/Parth Parsana/OneDrive/Desktop/Interview prep/Fetch reward/brands.csv',index=False)
```

## Converting Users JSON to readble CSV
